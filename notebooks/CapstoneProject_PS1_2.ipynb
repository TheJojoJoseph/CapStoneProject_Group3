## Overview on GANs (Generative Adversarial Networks)
GANs consist of two neural networks:
- Generator: Creates fake (synthetic) data.
- Discriminator: Tries to distinguish between real data and synthetic (fake) data.

The goal is to train the generator to create data that looks so real that the discriminator cannot reliably tell if it's fake. Over many training iterations, the generator improves and starts creating data that mimics the real data distribution very closely.

## Generating Synthetic Data for Augmentation Using GANs

- Selecting columns to generate synthetic data.
- Standardising the data for better GAN performance: Scaling the data to a standard normal distribution (mean=0, std=1) using 'StandardScaler' helps GANs converge more efficiently, especially since these columns have varying ranges (e.g., prices, deposit amounts).

```python
from sklearn.preprocessing import StandardScaler

gan_columns = knn_impute_cols
gan_data = final_data[gan_columns].dropna()  # Drop rows with NaN values for GAN input
scaler = StandardScaler()
gan_data_scaled = scaler.fit_transform(gan_data)
```

Defining the Generator


```python
import tensorflow as tf
from tensorflow.keras import layers

"""
The generator takes a random noise vector and transforms it into a synthetic data
sample with the same shape as the input data.

"""
def build_generator(latent_dim, output_dim):
    model = tf.keras.Sequential([
        layers.Dense(128, activation='relu', input_dim=latent_dim),
        layers.Dense(64, activation='relu'),
        layers.Dense(output_dim, activation='linear')
    ])
    return model

"""
latent_dim: The size of the input vector to the generator.
Itâ€™s a random noise vector that the generator uses to produce new samples.

output_dim: The number of features in the synthetic data (i.e., the number of columns we selected in gan_columns).

Generator Layers:
- Dense Layers: Each layer has a certain number of neurons (128, 64), with ReLU activation to introduce non-linearity and improve learning.
- Output Layer: Uses linear activation to produce outputs in a continuous range matching our scaled data.
"""
```

Defining the Discriminator

```python
"""
The discriminator tries to distinguish between real and fake samples created by the generator.
"""

def build_discriminator(input_dim):
    model = tf.keras.Sequential([
        layers.Dense(128, activation='relu', input_dim=input_dim),
        layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

"""
input_dim: The number of features in each sample (same as output_dim in the generator).

Discriminator Layers:
- Dense layers with ReLU activation.
- Output Layer: Uses sigmoid activation to produce a probability between 0 and 1, indicating whether a sample is real (1) or fake (0).
"""
```

Compiling the GAN

```python
latent_dim = 100
generator = build_generator(latent_dim, gan_data_scaled.shape[1])
discriminator = build_discriminator(gan_data_scaled.shape[1])

discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
discriminator.trainable = False

gan_input = layers.Input(shape=(latent_dim,))
gan_output = discriminator(generator(gan_input))
gan = tf.keras.Model(gan_input, gan_output)
gan.compile(optimizer='adam', loss='binary_crossentropy')

"""
latent_dim is set to 100 (the size of the random noise input for the generator).

Compile Discriminator: We use binary cross-entropy loss because itâ€™s a classification problem (real vs. fake), and adam optimizer for faster convergence.

discriminator.trainable = False: We "freeze" the discriminator when training the GAN as a whole, so the GANâ€™s training updates only the generatorâ€™s weights.

GAN Model: The GAN is created by connecting the generator and discriminator. The GAN input is noise, which goes through the generator to produce synthetic data. The discriminator then evaluates this data, outputting a probability score.
"""
```

Training the GAN

```python
"""
We alternate between training the discriminator (to improve its ability to detect fake samples)
and training the generator (to improve its ability to produce realistic samples).
"""

def train_gan(gan, generator, discriminator, data, epochs=1000, batch_size=128):
    for epoch in range(epochs):
        # Generate fake data
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        generated_data = generator.predict(noise)

        # Get a random batch of real data
        real_data = data[np.random.randint(0, data.shape[0], batch_size)]

        # Train discriminator
        combined_data = np.vstack((real_data, generated_data))
        labels = np.array([1] * batch_size + [0] * batch_size)
        discriminator.trainable = True
        d_loss = discriminator.train_on_batch(combined_data, labels)

        # Train generator
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        y_gen = np.ones((batch_size, 1))
        discriminator.trainable = False
        g_loss = gan.train_on_batch(noise, y_gen)

        if epoch % 100 == 0:
            print(f"Epoch {epoch}: D Loss = {d_loss[0]}, G Loss = {g_loss}")

train_gan(gan, generator, discriminator, gan_data_scaled, epochs=5000, batch_size=64)

"""
- Noise Generation: We create random noise input for the generator.
- Generator Training: Generates fake samples based on this noise.
- Real Data Sampling: Randomly selects real data samples of the same size as the batch.
- Combine Data and Train Discriminator: Combines real and fake data, labels them (1 for real, 0 for fake), and trains the discriminator.
- Train Generator via GAN: Freezes the discriminator and trains the GAN model, updating only the generatorâ€™s weights. The generator tries to "fool" the discriminator by making realistic samples.
- Epoch Logging: Every 100 epochs, the training process prints the discriminator and generator loss values to monitor progress.
"""
```

Generating Synthetic Data

```python
num_samples = 1000
noise = np.random.normal(0, 1, (num_samples, latent_dim))
synthetic_data = generator.predict(noise)

# Reverse scaling
synthetic_data = scaler.inverse_transform(synthetic_data)
synthetic_data_df = pd.DataFrame(synthetic_data, columns=gan_columns)
```

Integrating Synthetic Data

```python
augmented_data = pd.concat([final_data, synthetic_data_df], ignore_index=True)
```

# Categorical Features (Using SMOTE or Placeholder Values)

**SMOTE (Synthetic Minority Over-sampling Technique)** is a statistical technique primarily used to handle imbalanced datasets by creating synthetic samples of minority classes.
We will use SMOTE for categorical or binary classification columns if they are imbalanced.

For categorical features like summary, space, neighborhood_overview, etc., which contain substantial missing values, SMOTE can be applied to categorical data that serves as a target or classification variable. For most other categorical columns, we use placeholders.

Placeholder Values for Non-Target Categorical Data

```python
# Columns to fill with 'Unknown' or other placeholders
placeholder_cols = [
    'summary', 'space', 'neighborhood_overview', 'notes', 'transit', 'thumbnail_url',
    'medium_url', 'xl_picture_url', 'host_location', 'host_about', 'host_response_time',
    'host_neighbourhood', 'neighbourhood', 'license', 'zipcode'
]

for col in placeholder_cols:
    final_data[col].fillna('Unknown', inplace=True)

# Impute 'first_review' and 'last_review' with a placeholder date if necessary, else drop
# final_data['first_review'].fillna('1900-01-01', inplace=True)
# final_data['last_review'].fillna('1900-01-01', inplace=True)
```

Identifying and Applying SMOTE for Imbalanced Categorical Target

```python
from imblearn.over_sampling import SMOTE

X = final_data.drop(columns=['host_response_time'])  # Drop target variable
y = final_data['host_response_time']                 # Target variable

# Convert categorical variables in X to numerical format with one-hot encoding
X = pd.get_dummies(X, drop_first=True)
""" SMOTE requires numeric data, so we need to convert categorical variables in X to numeric using one-hot encoding.
One-hot encoding creates binary columns for each category, making it easier for SMOTE to generate synthetic samples."""

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

"""
- Initialize SMOTE with a random state for reproducibility.
- fit_resample: SMOTE generates synthetic samples by interpolating existing data points in the minority class,
creating a balanced dataset with X_resampled (resampled features) and y_resampled (resampled target).
"""

# Combine resampled data
resampled_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='host_response_time')], axis=1)
```

Combining Resampled Data Back into Original Data

```python
augmented_data = pd.concat([final_data, resampled_data], ignore_index=True)
```
